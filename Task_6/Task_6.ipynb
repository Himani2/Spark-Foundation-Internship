```python
#seasons which had maximum no.of matches
sns.countplot(x='season', data=df)
plt.show()
```

In 2013 most no.of matches are played

```python
data = df.winner.value_counts()
sns.barplot(y = data.index, x = data, orient='h')
```

Mumbai Indians are winning the most no.of matches

```python
#top players of the match are as follows
top_players = df.player_of_match.value_counts()[:10]
#sns.barplot(x="day", y="total_bill", data=df)
fig, ax = plt.subplots()
ax.set_ylim([0,20])
ax.set_ylabel("Count")
ax.set_title("Top player of the match Winners")
top_players.plot.bar()
sns.barplot(x = top_players.index, y = top_players, orient='v', palette="Blues");
plt.show()
```

CH Gayle is the most successful player.
